```python
#create model class has 1 neuron
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
  def __init__(self,num_features):#pass the total feature
    super().__init__()
    #input = num_features
    self.linear = nn.Linear(num_features,1)#input,output
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear(features) #we are calculating wx+b
    out = self.sigmoid(out)
    return out
```

```python
#dataset
features = torch.rand(10,5)

#create model
model = Model(features.shape[1])

#call forward pass
model(features)#standard way of calling insted of model.fordward()

#
```
